```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

file_path = '과제3 실시간 주문 대응 Routing 최적화 (od_matrix) 수정완료.csv'
od_matrix = pd.read_csv(file_path, encoding="cp949")
orders_table = pd.read_csv("과제3 실시간 주문 대응 Routing 최적화 (orders_table) 수정완료.csv", encoding="cp949")
terminals = pd.read_csv("과제3 실시간 주문 대응 Routing 최적화 (Terminals).csv", encoding="cp949")
veh_table = pd.read_csv("과제3 실시간 주문 대응 Routing 최적화 (veh_table).csv", encoding="cp949")
print(orders_table.loc[0]['터미널ID'])
edited_orders_node = {}
orders_list = {}
for i in range(0, len(orders_table)):
    if orders_table.loc[i]['터미널ID'] not in edited_orders_node:
        edited_orders_node[orders_table.loc[i]['터미널ID']]={}
    edited_orders_node[orders_table.loc[i]['터미널ID']][orders_table.loc[i]['주문ID']] = orders_table.loc[i]['터미널ID']+"_"+str(len(edited_orders_node[orders_table.loc[i]['터미널ID']]))
    orders_list[orders_table.loc[i]['터미널ID']] = [orders_table.loc[i]['터미널ID']+"_"+str(len(edited_orders_node[orders_table.loc[i]['터미널ID']]))]
print(edited_orders_node)
print(orders_list)
```
