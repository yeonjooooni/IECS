```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

file_path = '3    Routing  (od_matrix) .csv'
od_matrix = pd.read_csv(file_path, encoding="cp949")
orders_table = pd.read_csv("3    Routing  (orders_table) .csv", encoding="cp949")
terminals = pd.read_csv("3    Routing  (Terminals).csv", encoding="cp949")
veh_table = pd.read_csv("3    Routing  (veh_table).csv", encoding="cp949")
print(orders_table.loc[0]['ID'])
edited_orders_node = {}
orders_list = {}
for i in range(0, len(orders_table)):
    if orders_table.loc[i]['ID'] not in edited_orders_node:
        edited_orders_node[orders_table.loc[i]['ID']]={}
    edited_orders_node[orders_table.loc[i]['ID']][orders_table.loc[i]['ID']] = orders_table.loc[i]['ID']+"_"+str(len(edited_orders_node[orders_table.loc[i]['ID']]))
    orders_list[orders_table.loc[i]['ID']] = [orders_table.loc[i]['ID']+"_"+str(len(edited_orders_node[orders_table.loc[i]['ID']]))]
print(edited_orders_node)
print(orders_list)
```
